```python
# Select numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64'])

# Mean
print("Mean")
print(numerical_columns.mean())

print("\nMedian")
print(numerical_columns.median())

print("\nMode")
print(numerical_columns.mode().iloc[0])

print("\nStandard Deviation")
print(numerical_columns.std())
```

```python
# Convert Order Date to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Order Date"].head()
```

```python
df["Month"] = df["Order Date"].dt.to_period("M")
df["Quarter"] = df["Order Date"].dt.to_period("Q")
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
quarterly_sales = df.groupby("Quarter")["Sales"].sum()

plt.figure(figsize=(8,5))
quarterly_sales.plot(marker='o', color='green')
plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Top 10 products by total sales
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_products.plot(kind='bar', color='skyblue')
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color=['steelblue','orange','green'])
plt.title("Revenue by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.select_dtypes(include=['int64','float64']).corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))
region_sales.plot(kind='barh', color='purple')

plt.title("Sales by Region")
plt.xlabel("Total Sales")
plt.ylabel("Region")
plt.show()
```

```python
# Business Insights

### Insight 1
#Technology products generate the highest revenue. Increasing inventory and promotional offers for this category can further boost sales.

### Insight 2
#Sales vary across different months and quarters, indicating seasonal demand. Planning marketing campaigns before peak seasons can improve revenue.

### Insight 3
#Regional sales performance differs significantly. High-performing regions should receive additional stock, while low-performing regions should be targeted with localized marketing strategies.
```
